```python
from deepfake_detection.defaults import DRIVE_PATH, DATA_PATH
import pandas as pd
from pathlib import Path
```

```python
metadata_df = pd.read_csv(DRIVE_PATH / "metadata.csv")
metadata_df.set_index("video", inplace=True)
metadata_df
```

```python
import os 

images_df = pd.read_json(DATA_PATH / "exported_images.json", orient="index")
images_df.reset_index(inplace=True)
images_df['index'] = images_df['index'].apply(lambda x: x.split("\\")[-1])
images_df = images_df[images_df[0].notnull() & images_df[1].notnull()]
images_df[0] = images_df[0].apply(lambda x: "/".join(x.split("\\")[1:]))
images_df[1] = images_df[1].apply(lambda x: "/".join(x.split("\\")[1:]))
images_df.set_index("index", inplace=True)
images_df
```

```python
merged_df = metadata_df.merge(images_df, left_index=True, right_index=True)
merged_df['image_path'] = merged_df[[0, 1]].values.tolist()
merged_df.drop([0, 1], axis=1, inplace=True)
merged_df = merged_df.explode('image_path')
merged_df.reset_index(inplace=True)
merged_df.rename(columns={"index": "video"}, inplace=True)
merged_df.set_index("image_path", inplace=True)
merged_df
```

```python
merged_df.to_csv(DATA_PATH / "images_data.csv")
```
